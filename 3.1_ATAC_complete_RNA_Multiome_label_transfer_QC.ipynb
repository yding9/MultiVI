### Compare subclass frequency pre and post filtering

```python
adata_mvi = anndata.read_h5ad(filename="multivi_AD_3cohorts_RNA_ATAC_Multiome_v1.2_trained_addQC.h5ad")
```

```python
df_pre = adata_mvi.obs
df_pre
```

```python
df_pre["rna_neighbors_qc_ratio"].isna().sum()
```

```python
ad = adata_mvi[adata_mvi.obs["rna_neighbors_qc_ratio"].to_numpy() < 0.01]
```

```python
# ad = anndata.read_h5ad(filename="multivi_AD_3cohorts_RNA_ATAC_Multiome_v1.3_trained.h5ad")
df_post = ad.obs
df_post
```

```python
del adata_mvi
del ad
```

```python
df_pre.reset_index(inplace=True)
df_pre
```

```python
df_pre = df_pre.rename(columns = {'level_0':'sample_index'})
df_pre
```

```python
df_post.reset_index(inplace=True)
df_post = df_post.rename(columns = {'level_0':'sample_index'})
df_post
```

```python
df_flt = df_pre[~df_pre.sample_index.isin(df_post.sample_index)]
```

```python
df_flt.groupby("modality").size()
```

```python
df_flt_rna = df_flt[df_flt.modality.isin(["paired", "expression"])]
```

```python
df_pre_rna = df_pre[df_pre.modality.isin(["paired", "expression"])]
```

```python
# % of good RNA/Multiome cells get removed in rna_neighbors_qc_ratio
print(np.sum(df_flt_rna["Doublet_or_LowQuality"] == "cells with high quality") / np.sum(df_pre_rna["Doublet_or_LowQuality"] == "cells with high quality"))
```

```python
subclass_ratio_pre = df_pre_rna.groupby("confirmed_subclass_scANVI").size() / df_pre_rna.shape[0]
```

```python
np.sum(subclass_ratio_pre)
```

```python
subclass_ratio_pre
```

```python
subclass_ratio_post = df_post[df_post.modality.isin(["paired", "expression"])].groupby("confirmed_subclass_scANVI").size() / df_post[df_post.modality.isin(["paired", "expression"])].shape[0]
```

```python
np.sum(subclass_ratio_post)
```

```python
subclass_ratio_post
```
